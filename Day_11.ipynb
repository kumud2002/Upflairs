```python
vid = cv.VideoCapture(0)
while True:
    flag , img = vid.read()
    if flag:
        cv.imshow('Preview',img)
        key = cv.waitKey(1)
        if key == ord('c'):
            break
    else:
        break
vid.release()
cv.destroyAllWindows()
```

Important - 
- vid.release()
- cv.destroyAllWindows()
- To release camera acquired and destroy windows

```python
# IN GRAY SCALE --

vid = cv.VideoCapture(0)
while True:
    flag , img = vid.read()
    if flag:
        # Processing Code (image came here)
        img_gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)      
        cv.imshow('Preview',img_gray)
        key = cv.waitKey(1)
        if key == ord('c'):
            break
    else:
        break
vid.release()
cv.destroyAllWindows()
```

```python
# Cropped Image --

vid = cv.VideoCapture(0)
while True:
    flag , img = vid.read()
    if flag:
        # Processing Code (image came here)
        x,y,w,h = (200,50,350,300)
        img_crop = img[y:y+h,x:x+w,:]  
        cv.rectangle(img,pt1 = (x,y) ,pt2 = (x+w,y+h),color = (0,0,255),thickness = 8)  
        cv.imshow('Preview',img)
        key = cv.waitKey(1)
        if key == ord('c'):
            break
    else:
        break
vid.release()
cv.destroyAllWindows()
```
