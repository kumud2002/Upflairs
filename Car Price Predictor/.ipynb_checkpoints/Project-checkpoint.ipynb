# Data Preparation

```python
x = df.drop(['price','stroke','carheight','compressionratio'],axis=1)
y = df['price']
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.2,random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression().fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)*100
```

```python
print('Model Score on Training Data: {}%'.format(round(model.score(xtrain,ytrain)*100,2)))
```

```python
print('Model Score on Testing Data: {}%'.format(round(model.score(xtest,ytest)*100,2)))
```

```python
pd.to_pickle(model, 'CarPricePredictor.pkl')
```

```python
wb = float(input('Wheelbase:          '))
cl = float(input('Carlength:          '))
es = int(input('Enginesize:         '))
cw = float(input('Carwidth:           '))
cbw = int(input('Curbweight:         '))
br = float(input('Boreratio:          '))
hp = int(input('Horsepower:         '))
querydf = pd.DataFrame({'Wheelbase': [wb], 'Carlength': [cl], 'Enginesize': [es], 'Carwidth': [cw],
                        'Curbweight': [cbw], 'Boreratio': [br], 'Horsepower': [hp]})
querydf
```

# GUI Devlopement

```python
import tkinter as tkn
```

```python
app = tkn.Tk()
app.geometry('450x350')
app.title('Price Predictor')
mod = pd.read_pickle('CarPricePredictor.pkl')

wbs = tkn.Variable(app)
crl = tkn.Variable(app)
egs = tkn.Variable(app)
crw = tkn.Variable(app)
crbw = tkn.Variable(app)
bor = tkn.Variable(app)
hrp = tkn.Variable(app)
result = tkn.Variable(app)

tkn.Label(app,text='Wheelbase:',font=('Century Gothic',15),fg='purple').place(x=10,y=20)
tkn.Label(app,text='Carlength:',font=('Century Gothic',15),fg='purple').place(x=10,y=50)
tkn.Label(app,text='Enginesize:',font=('Century Gothic',15),fg='purple').place(x=10,y=80)
tkn.Label(app,text='Carwidth:',font=('Century Gothic',15),fg='purple').place(x=10,y=110)
tkn.Label(app,text='Curbweight:',font=('Century Gothic',15),fg='purple').place(x=10,y=140)
tkn.Label(app,text='Boreratio:',font=('Century Gothic',15),fg='purple').place(x=10,y=170)
tkn.Label(app,text='Horsepower:',font=('Century Gothic',15),fg='purple').place(x=10,y=200)
tkn.Label(app,textvariable=result,font=('Baskerville Old Face',15),fg='blue').place(x=150,y=290)

tkn.Entry(app,textvariable=wbs,width=15,font=('Franklin Gothic Book',15),bg='tan').place(x=170,y=20)
tkn.Entry(app,textvariable=crl,width=15,font=('Franklin Gothic Book',15),bg='tan').place(x=170,y=50)
tkn.Entry(app,textvariable=egs,width=15,font=('Franklin Gothic Book',15),bg='tan').place(x=170,y=80)
tkn.Entry(app,textvariable=crw,width=15,font=('Franklin Gothic Book',15),bg='tan').place(x=170,y=110)
tkn.Entry(app,textvariable=crbw,width=15,font=('Franklin Gothic Book',15),bg='tan').place(x=170,y=140)
tkn.Entry(app,textvariable=bor,width=15,font=('Franklin Gothic Book',15),bg='tan').place(x=170,y=170)
tkn.Entry(app,textvariable=hrp,width=15,font=('Franklin Gothic Book',15),bg='tan').place(x=170,y=200)

def prediction():
    global mod
    wheelbase = eval(wbs.get())
    carlength = eval(crl.get())
    enginesize = eval(egs.get())
    carwidth = eval(crw.get())
    curbweight = eval(crbw.get())
    boreratio = eval(bor.get())
    horsepower = eval(hrp.get())
    query = [[wheelbase, carlength, enginesize, carwidth, curbweight, boreratio, horsepower]]
    result.set(round(mod.predict(query)[0]*76.013/100000,1))
    tkn.Label(app,text='(in Lakhs)',font=('Baskerville Old Face',15),fg='red').place(x=250,y=290)

tkn.Button(app, text='Predict Price', font=('Castellar',15),bg='aqua', command=prediction).place(x=150,y=240)

app.mainloop()
```
